### ДЗ 3.1. Поиск с помощью BLAST
#### Буклей Григорий

Задание: Напишите генератор случайной последовательности нуклеотидов. Сгенерируйте с помощью него 3 случайные последовательности (длины 100, 1000, 100 000).

 С помощью инструмента blastn (https://blast.ncbi.nlm.nih.gov/Blast.cgi) проведите поиск по полученным последовательностям по “nucleotide collection (nr/nt)

```python
import random

def generate_query(length, file):
    f = open(file+'.txt', 'w')
    array = [random.randint(0, 3) for i in range(length)]
    for el in array:
        if el == 0:
            f.write('A')
        elif el == 1:
            f.write('C')
        elif el == 2:
            f.write('G')
        elif el == 3:
            f.write('T')
    f.close()
```

```python
generate_query(100, "random_query_100")
generate_query(1000, "random_query_1000")
generate_query(100000, "random_query_100000")
```
